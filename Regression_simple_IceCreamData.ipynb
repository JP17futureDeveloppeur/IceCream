```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('IceCreamData.csv')
data.head()
```

```python
sns.pairplot(data)
```

```python
#Séparer X et y
X = data.drop(['Revenue'], axis=1)
y = data [['Revenue']]
```

```python
#Séparer X_train,X_test,y_train,y_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
#Appeler et entrainer le modèle de régression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Montrer les coefficients
print('a =',model.coef_)
print('b =',model.intercept_)
```

```python
#Train
#Prediction du modèle
from sklearn.metrics import mean_squared_error
y_pred_train = model.predict(X_train)
#Evaluation du modèle
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('Mean Squared Error train:', mean_squared_error(y_train,y_pred_train))
print('R2 score train:', r2_score(y_train,y_pred_train))

```

```python
#Test
#Prediction du modèle
from sklearn.metrics import mean_squared_error
y_pred_test = model.predict(X_test)
#Evaluation du modèle
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('Mean Squared Error test:', mean_squared_error(y_test,y_pred_test))
print('R2 score test:', r2_score(y_test,y_pred_test))
```

```python
#Simulation de prédiction
y_predict = model.predict([[30]])
print('Prediction:',y_predict)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,model.predict(X_train),color='blue')
plt.title('Revenue vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
```
